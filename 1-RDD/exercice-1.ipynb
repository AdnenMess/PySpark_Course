```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("TotalAmount")
sc = SparkContext(conf=conf)
```

```python
input = sc.textFile("Data/customer-orders.csv")
```

```python
def parseLine(line):
    field = line.split(',')
    customerID = int(field[0])
    amout = float(field[2])
    return customerID,amout
```

```python
rdd = input.map(parseLine)
```

```python
totalAmount = rdd.reduceByKey(lambda x,y : x+y)
```

```python
totalAmountSorted = totalAmount.map(lambda x : (x[1], x[0])).sortByKey(False).map(lambda x : (x[1], x[0]))
```

```python
result = totalAmountSorted.collect()
```

```python
for total in result:
    print(total[0],' ==> ', total[1])
```
